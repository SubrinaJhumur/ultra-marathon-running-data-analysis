```python
# importing necessary libraries for the project
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# data loading
data=pd.read_csv(r"C:\Users\User\Desktop\TWO_CENTURIES_OF_UM_RACES.csv")
```

```python
# first 5 rows of the dataset
data.head()
```

```python
# total number of rows and columns in the dataset
data.shape
```

```python
# dataset description
data.info()
```

```python
# clean up data
# want only usa data
new_data=data[(data['Event distance/length']=='50km')| (data['Event distance/length']=='50mi')]
```

```python
# new dataset only for 2020 year
new_data_1=new_data[new_data['Year of event']==2020]
```

```python
# showing new dataset head
new_data_1.head()
```

```python
# extracting country from Event name column
new_data_1["Country"] = new_data_1["Event name"].str.extract(r"\((.*?)\)")
```

```python
# new dataset head
new_data_1.head()
```

```python
# taking countries only in USA
df=new_data_1[new_data_1["Country"]=='USA']
```

```python
# showing the new dataset
df
```

```python
# removing country name from the Event name column
df["Event name"] = df["Event name"].str.replace(r"\s*\(.*?\)", "", regex=True)
```

```python
# creating new column of Athlete_Age from the dataset
df['Athlete_Age']=2020-df['Athlete year of birth']
```

```python
df.head()
```

```python
# removing h from Athlete_Age column
df["Athlete_performance"] = data["Athlete performance"].str.replace(" h", "", regex=False)
```

```python
df.head()
```

```python
# drop columns: Athlete club, Athlete country,Athlete year of birth,Athlete age category
df.drop(columns={'Athlete club','Athlete country','Athlete year of birth','Athlete age category'},inplace=True)
```

```python
# clean up null values
df.isnull().sum()
```

```python
df=df.dropna()
```

```python
# new shape of the dataset
df.shape
```

```python
# check for duplicates
df.duplicated().sum()
```

```python
# reset index
df = df.reset_index()
```

```python
df.head(3)
```

```python
# fix types
df.dtypes
```

```python
# change the data type of Athlete_Age column
df['Athlete_Age']=df['Athlete_Age'].astype('int')
```

```python
# change the data type of Athlete average speed column
df['Athlete average speed']=df['Athlete average speed'].astype('float')
```

```python
# rename columns
#index                          int64
##Event dates                   object
#Event name                    object
#Event distance/length         object
#Event number of finishers      int64
#Athlete performance           object
#Athlete gender                object
#Athlete average speed        float64
#Athlete ID                     int64
#Country                       object
#Athlete_Age                    int32
#Athlete_performance           object
```

```python
df=df.rename(columns={'Year of event':'Year',
                      'Event dates':"race_day",
                      "Event name":"race_name",
                     "Event distance/length":'race_distance',
                     "Event number of finishers":"race_number_of_finishers",
                     "Athlete performance":'Athlete_performance',
                     "Athlete gender":'Athlete_gender',
                     "Athlete average speed":"Athlete_average_speed",
                     "Athlete ID":'Athlete_ID',
                        })
```

```python
df.head()
```

```python
# reorder columns
df1=df[['index','Athlete_ID', 'Country','Year', 'Athlete_Age','race_name', 'race_day',  'race_distance','race_number_of_finishers', 'Athlete_performance', 'Athlete_gender','Athlete_average_speed', 'Athlete_performance']]
```

```python
df1.head()
```

```python
# check 222509 from Athlete_ID column
df1[df1['Athlete_ID']==222509]
```

```python
# charts and graphs
sns.histplot(df1['race_distance'])
plt.show()
```

```python
# race_distance histogram plot
sns.histplot(df1, x='race_distance',hue='Athlete_gender')
plt.show()
```

```python
# showing the df1['race_distance']=='50mi' graph
sns.displot(df1[df1['race_distance']=='50mi']['Athlete_average_speed'])
plt.show()
```

```python
# showing the violin plot for male and female for race_distance and Athlete_average_speed
sns.violinplot(data=df1,x='race_distance',y='Athlete_average_speed',hue='Athlete_gender',split=True,inner='quart')
```

```python
# showing lmplot for male and female in Athlete_Age and Athlete_average_speed chart
sns.lmplot(data=df1,x='Athlete_Age',y='Athlete_average_speed',hue='Athlete_gender')
plt.show()
```

```python
# questions i want to find out from the data
# difference in speed in 50 km and 50 mi in male and female
df1.groupby(['race_distance','Athlete_gender'])['Athlete_average_speed'].mean()
```

```python
# what age groups are the best in 50 mi race(20+ races min)
df1.query("race_distance =='50mi'").groupby('Athlete_Age')['Athlete_average_speed'].agg(['mean','count']).sort_values('mean',ascending=False).query('count>19').head(15)
```

```python
# what age groups are the worst in 50 mi race(10+ races min)(show 20)
df1.query("race_distance =='50mi'").groupby('Athlete_Age')['Athlete_average_speed'].agg(['mean','count']).sort_values('mean',ascending=True).query('count>9').head(20)
```

```python
# seasons for the data
# spring 3-5
# summer 6-8
# fall 9-11
# winter 12-2
# split between 2 decimals
df1['race_month']=df1['race_day'].str.split('.').str.get(1).astype(int)
```

```python
df1.head()
```

```python
# season extracting from the dataset
df1['race_season']=df1['race_month'].apply(lambda x:'Winter' if x>11 else 'Fall' if x>8 else 'Summer' if x>5 else 'Spring' if x>2 else 'Winter' )
```

```python
# showing how many events happened in each season 
df1['race_season'].value_counts()
```

```python
# sorting the Athlete_average_speed in acording to the race_season
df1.groupby('race_season')['Athlete_average_speed'].agg(['mean','count']).sort_values('mean',ascending=False)
```

```python
# sorting the Athlete_average_speed in acording to the race_season for 50mi only
df1.query("race_distance=='50mi'").groupby('race_season')['Athlete_average_speed'].agg(['mean','count']).sort_values('mean',ascending=False)
```
